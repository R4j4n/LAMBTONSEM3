```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_vectors(vectors):
 
    # Set up the 3D plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    colors = ['r', 'b', 'g', 'purple', 'orange', 'cyan']
    
    origin = np.zeros(3)
    
    for i, vector_str in enumerate(vectors):
        components = {'a': 0, 'b': 0, 'c': 0}
        terms = vector_str.replace('-', '+-').split('+')
        
        for term in terms:
            if term:
                if term.endswith('a'):
                    components['a'] = float(term[:-1] if term[:-1] else 1)
                elif term.endswith('b'):
                    components['b'] = float(term[:-1] if term[:-1] else 1)
                elif term.endswith('c'):
                    components['c'] = float(term[:-1] if term[:-1] else 1)
        
        vector = np.array([components['a'], components['b'], components['c']])
        
        color = colors[i % len(colors)]  # Cycle through colors if more vectors than colors
        ax.quiver(origin[0], origin[1], origin[2],
                 vector[0], vector[1], vector[2],
                 color=color, label=f'Vector {i+1}: {vector_str}')
    
    ax.set_xlabel('a')
    ax.set_ylabel('b')
    ax.set_zlabel('c')
    ax.set_title('3D Vector Visualization')
    ax.legend()
    ax.grid(True)
    ax.set_box_aspect([1,1,1])
    plt.show()
```

```python
def parse_vector(vector_str):

    components = {'a': 0, 'b': 0, 'c': 0}
    # Handle negative signs by replacing - with +- for splitting
    terms = vector_str.replace('-', '+-').split('+')
    
    for term in terms:
        if term:  # Skip empty terms
            # Remove spaces if any
            term = term.strip()
            if term.endswith('a'):
                components['a'] = float(term[:-1] if term[:-1] else 1)
            elif term.endswith('b'):
                components['b'] = float(term[:-1] if term[:-1] else 1)
            elif term.endswith('c'):
                components['c'] = float(term[:-1] if term[:-1] else 1)
    
    return components

def dot_product(vec1, vec2):

    # Parse both vectors
    v1 = parse_vector(vec1)
    v2 = parse_vector(vec2)
    
    # Calculate dot product
    dot_prod = (v1['a'] * v2['a'] + 
                v1['b'] * v2['b'] + 
                v1['c'] * v2['c'])
    
    return dot_prod
```

```python
vec1 = '10a+20b+30c'
vec2 = '5a-20b+45c'

vec3 = dot_product(vec1,vec2)

plot_vectors([vec1,vec2])

print(f"Dot product: {vec3}")
```
